## OOPS in Python

### Concept of Class & Object

1) Consider a Vehicle class with multiple objects

```python
class Vehicle:
    """
    class is defined here with keywork 'class' and class_name with first letter always capital.
    Dundar method of python, self catches the object, everytime it recieves an object.
    A "constructor" is called when we create an object.
    
    """
    def __init__(self):
        print("Vehicle default constructor")
        
    def __init__(self, name):
        self.name = name # a parameter called "name" is defined here
        print("Vehicle parameterised cunstructor")

# create different functions that a vehicle can perform, 
#e.g. ride, clean, accelerate, brake, etc.

    def ride(self): 
        print("riding vehicle")
        
    def clean(self):
        print("Vehicle cleaning")

object1 = Vehicle("Car") # object creation for class vehicle
print(object1.name)
object1.clean()        
object1.ride()   
```

```python
# another object can be created
object2 = Vehicle("Bike")
print(object2.name)
object2.clean()        
object2.ride()
```

### Inheritance

Consider a bass class Animal and derived classes namely Human and Dog

```python
class Animal:  # Base class
    
    def __init__(self, name):
        self.any_name = name
    
    def eat(self,food):
        self.food = food
        print("method overriding")

class Human(Animal): # Derived class of Animal base class
    
    def __init__(self, name):
        self.naam = name

class Dog(Animal):  # Derived class of Animal base class
    
    def __init__(self, name):
        self.naam = name
        
human = Human("Guzayya") # Human object created for Human derived class
# eat method is defined in Animal base class and called by human object,
#hence this is called "function overriding"
human.eat("Biryani") # Inherited function "eat" is called "overridden function"
print("*********")
dog = Dog("puppy") # Dog object created for Human derived class
dog.eat("dog food") 
```

## Nested Loop

The traditional Star pattern using python nested for loop

```python
n = 6
for row in range(1,n):  # represents number of rows
    for col in range(row): # represents number of columns
        print("*", end=" ")
    print("\r")
```

Same star pattern using nested while loop

```python
row = 6
while(row>0):
    col = 6
    while(col > row):
        print("*",end=" ")
        col-=1
    row-=1
    print()
```

### Loop control statements in Python

### continue, break, pass

#### break

```python
for alphabet in "python":
    print(alphabet)
    if (alphabet == "h"):
        break   #breaks the loop
```

#### continue

```python
num = 0
while(num<6):
    num +=1
    if num == 4:
        continue # control goes back to the iterating loop
    print(num)
```

#### pass

```python
for num in range(5):
    pass # does nothing(null statement), performs No operation
print(num)
```

## Recursion

Simple example to calculate the sum of series

```python
def sum_function(num):
    total_sum = 0
    for index in range(num + 1):
        total_sum += index
    return total_sum

result = sum_function(10)
print(result)
```

Recursive function calls itself. e.g: below quick sort function(covered in Week 3)

```python
import numpy as np
def quick_sort(numpy_array):

    left_element = []
    equal_element = []
    right_element = []

    if len(numpy_array) > 1:
        pivot_value = numpy_array[0]
        for num in numpy_array:
            if num < pivot_value:
                left_element.append(num)
            elif num == pivot_value:
                equal_element.append(num)
            elif num > pivot_value:
                right_element.append(num)

        return quick_sort(left_element) + equal_element + quick_sort(right_element)
        #recursively call quick_sort() function
# If there exists only 1 element, then return the array    
    else:  
        return numpy_array
quick_sort(np.array([4,3,5,6]))
```

### Exception handling via try - except block

```python
import sys
list1 = ["A", 0, 2]

for element in list1:
    try:
        print(f"The element is : {element}")
        reciprocal = 1/int(element)
        break
    except:
        print(f"Found {sys.exc_info()[0]}")  #sys.exc_info() is a system method which returns a tuple of 
        #(type, value, traceback), where type gives exception name e.g. ValueError, NameError, AttributeError, etc.
        print(f"Try next element")
        print()
print(f"Reciprocal of {element} is : {reciprocal}")
```

## Debugging in Python

### breakpoint() method

Commands for debugging:

|command | description |
| --- | --- |
| c | continue execution |
| q | quit |
| n | jumps to next line in the same function|

Example 1:

```python
def debug_function(num1, num2):
    breakpoint()
    division = num1 / num2
    return division
  
debug_function(5, 0)
```

Example 2:

```python
def debug_function(num_list):
    breakpoint()
    output = [num_list[value] for value in range(len(num_list) + 4)]
    return output
debug_function([2, 4, 10, 77])
```

### pdb in-build library

```python
import pdb
def debug_function(num1, num2):
    pdb.set_trace()
    division = num1 / num2
    return division

debug_function(5, 0)
```

## Optimization in Python

- Use list comprehensions wherever possible
- Avoid unnecessary functions as they are expensive(memory)
- Use build-in functions whenever possible
- Assign variables to object methods ( sum = list.sum )
- Know your python data structures and how they behave in your installed version
- and many more

Example 1: Use of list comprehension

```python
num_list = [1,2,3,4,5,6]
even_num_list = []
for num in num_list:
    if (num % 2) == 0:
        even_num_list.append(num)
print(f"{even_num_list}")
```

Optimized using list comprehension

```python
even_num_list = [num for num in num_list if num % 2 == 0]
even_num_list
```

Example 2: Use of build in functions

```python
string_list = ["Python ","is ","a ","beautiful ","language"]
string_output = ""
for element in string_list:
    string_output += element
string_output
```

Optimized using build in str.join() method of string

```python
string_output = ''.join(string_list)
string_output
```
