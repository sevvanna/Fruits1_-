### Sharpen Filter on the Image

```python
from matplotlib.pyplot import imread
from scipy.ndimage.filters import convolve

sharpen = np.array([ [0, -1, 0], [-1, 5, -1], [0, -1, 0]])   
img = imread("/content/Dog.jpg")
Image_channel = []

for channel in range(3):
    res = convolve(img[:,:,channel], sharpen)
    Image_channel.append(res)

img = np.dstack((Image_channel[0], Image_channel[1], Image_channel[2]))
print(img.shape)
plt.imshow(img)
plt.show()
```

### Edge detection filter on the image.

```python
Edge = np.array([ [0, 1, 0], [1,-8,1], [0, 1, 0]])  
img = imread("/content/Dog.jpg")
Image_channel = []

for channel in range(3):
    res = convolve(img[:,:,channel], Edge)
    Image_channel.append(res)

img = np.dstack((Image_channel[0], Image_channel[1], Image_channel[2]))
plt.imshow(img)
plt.show()
```

### Filtering Images using ImageFilter module.

```python
# Import required image modules
from PIL import Image, ImageFilter

# Filter import
from PIL.ImageFilter import (BLUR, CONTOUR, DETAIL, EDGE_ENHANCE, EDGE_ENHANCE_MORE,
   EMBOSS, FIND_EDGES, SMOOTH, SMOOTH_MORE, SHARPEN)


# Create a list of filters
filters = [BLUR, CONTOUR, DETAIL, EDGE_ENHANCE, EDGE_ENHANCE_MORE, EMBOSS, 
           FIND_EDGES, SMOOTH, SMOOTH_MORE, SHARPEN]

for i in filters:
    img = Image.open('/content/Dog.jpg')
    img1 = img.filter(i)  # choose different filters from above
    plt.imshow(img1)
    plt.show()
```
